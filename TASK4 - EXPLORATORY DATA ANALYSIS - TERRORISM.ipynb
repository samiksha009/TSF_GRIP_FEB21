```python
#import required packages and libraries

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
#import and load the dataset

df = pd.read_csv("globalterrorism.csv", encoding="ISO-8859-1")
df
```

```python
#checking for null values

df.isnull().sum()
```

```python
#check the duplicates

df.duplicated().sum()
```

```python
df.dropna(axis=1, how="all")
```

```python
#rename the columns which aren't well labelled 

df.rename(columns={"iyear" : "Year", "imonth" : "Month", "iday" : "Day", "country_txt" : "Country",
                   "region_txt" : "Region", "provstate" : "State", "latitude" : "Latitude", 
                   "longitude" : "Longitude", "summary" : "Summary", "attacktype1_txt" : "AttackType", 
                   "targtype1_txt" : "TargetType", "gname" : "Group", "weaptype1_txt" : "WeaponType", 
                   "nkill" : "Kill", "nwound" : "Wound",}, inplace=True)
```

```python
df
```

```python
#keeping only the columns which can be responsible for terrorist attacks and which should be taken into account 
df = df[["Year", "Month", "Day", "AttackType", "TargetType", "Country", "Region",
         "Latitude", "Longitude", "Summary", "State", "motive", "WeaponType", 
         "Kill", "Wound", "Group", "Summary"]]
df
```

## EXPLORATORY DATA ANALYSIS

```python
print("Year with the most terrorist attacks :", df["Year"].value_counts().idxmax())
print("Month with the most terrorist attacks :", df["Month"].value_counts().idxmax())
print("Most terrorist attacks types :", df["AttackType"].value_counts().idxmax())
```

```python
print("Country with the highest terrorist attacks :", df["Country"].value_counts().idxmax())
```

```python
print("Country with the highest terrorist attacks :", df["Country"].value_counts().idxmax())
print("Region with the highest terrorist attacks :", df["Region"].value_counts().idxmax())
print("Group with the most terrorist attacks :", df["Group"].value_counts().index[0])
```

```python
# since Unknown is something ths is not known, we should try this

print("Group with the most terrorist attacks :", df["Group"].value_counts().index[1])
```

## Various attacking methods

```python
plt.figure(figsize=(20,10))
sns.countplot(x="AttackType", data=df, order = df["AttackType"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Various attacking methods")
plt.show()
```

## What are the major terrorist targets? 

```python
plt.figure(figsize=(20,10))
sns.countplot(x="TargetType", data=df, order = df["TargetType"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Major Terrorist Targets")
plt.show()
```

## Which regions are the most affected terrorist targets?

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Region", data=df, order = df["Region"].value_counts().index)
plt.xticks(rotation=90)
plt.title("Most affected terrorist targets region wise")
plt.show()

```

## Which countries are the most affected countries by terrorism? 

```python
plt.figure(figsize=(20,10))

x=df["Country"].value_counts().index
y=df["Country"].value_counts().values
sns.barplot(x, y)
plt.title("Countries affected by Terrorism")
plt.show()
```

```python
# since the countries are not visible, let's do it till the top 15 countries

plt.figure(figsize=(20,10))
x=df["Country"].value_counts()[:15].index
y=df["Country"].value_counts()[:15].values
sns.barplot(x, y)
plt.title("Countries affected by Terrorism")
plt.show()
```

## No. of people got killed country wise 

```python
plt.figure(figsize=(20,10))

x=df["Country"].value_counts()[:15].index
y=df["Kill"].value_counts()[:15].values
sns.barplot(x, y)
plt.title("Countries with no. of people killed in terrorist attacks")
plt.show()
```

# No. of terrorist attacks w.r.t year

```python
plt.figure(figsize=(20,10))

x=df["Year"].value_counts().index
y=df["Year"].value_counts().values
sns.barplot(x, y)
plt.show()
```
